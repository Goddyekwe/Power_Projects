```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.multioutput import MultiOutputClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
from sklearn.preprocessing import StandardScaler

# Split input features and target labels
X = df.drop(['Fault_Class', 'Line_Faulty', 'Fault_Location'], axis=1)
y = df[['Fault_Class', 'Line_Faulty', 'Fault_Location']]
```

```python

# Encode target labels using LabelEncoder
encoders = {}
y_encoded = pd.DataFrame()

for col in y.columns:
    le = LabelEncoder()
    y_encoded[col] = le.fit_transform(y[col])
    encoders[col] = le  # Save encoder for inverse transform later
```

```python
# Train/test split
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=101)
```

```python
# Scale the input features (X)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# Define the base classifier and wrap it in MultiOutputClassifier
base_model = RandomForestClassifier(n_estimators=100, random_state=101)
multi_output_model = MultiOutputClassifier(base_model)
```

```python
# Train the model
multi_output_model.fit(X_train_scaled, y_train)
```

```python

# Predict on test data
y_pred = multi_output_model.predict(X_test_scaled)

```

```python
# Evaluate each target separately
for i, col in enumerate(y_encoded.columns):
    print(f"\n🟩 Classification Report for '{col}':\n")
    print(classification_report(y_test.iloc[:, i], y_pred[:, i], target_names=encoders[col].classes_))
```

```python
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

for i, col in enumerate(y_encoded.columns):
    cm = confusion_matrix(y_test.iloc[:, i], y_pred[:, i])
    plt.figure(figsize=(18, 9))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Greys', xticklabels=encoders[col].classes_, yticklabels=encoders[col].classes_)
    plt.title(f'Confusion Matrix for {col}')
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.show()
```

```python
from sklearn.metrics import accuracy_score

for i, col in enumerate(y_encoded.columns):
    acc = accuracy_score(y_test.iloc[:, i], y_pred[:, i])
    print(f"Accuracy for {col}: {acc:.3f}")
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score

# Initialize lists to store metrics for each target column
precision_scores = []
recall_scores = []
f1_scores = []

# Calculate metrics for each target column
for i, col in enumerate(y_test.columns):
	precision = precision_score(y_test.iloc[:, i], y_pred[:, i], average='macro')
	recall = recall_score(y_test.iloc[:, i], y_pred[:, i], average='macro')
	f1 = f1_score(y_test.iloc[:, i], y_pred[:, i], average='macro')
	
	precision_scores.append(precision)
	recall_scores.append(recall)
	f1_scores.append(f1)
	
	print(f"\nMetrics for '{col}':")
	print(f"  Precision: {precision:.2f}")
	print(f"  Recall: {recall:.2f}")
	print(f"  F1-Score: {f1:.2f}")

# Optionally, calculate the average across all target columns
avg_precision = sum(precision_scores) / len(precision_scores)
avg_recall = sum(recall_scores) / len(recall_scores)
avg_f1 = sum(f1_scores) / len(f1_scores)

print("\nOverall Macro-Averaged Metrics:")
print(f"  Precision: {avg_precision:.2f}")
print(f"  Recall: {avg_recall:.2f}")
print(f"  F1-Score: {avg_f1:.2f}")
```
